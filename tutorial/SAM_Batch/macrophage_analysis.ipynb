# Activated macrophage analysis

In this tutorial, I demonstrate the use of `calculate_regression_PCs` and `regress_genes` to remove cell-cycle effects from the data. This data is taken from the following paper:

Lane, Keara, et al. "Measuring signaling and RNA-seq in the same cell links gene expression to dynamic patterns of NF-ÎºB activation." Cell systems 4.4 (2017): 458-469.

First, we import the necessary packages:

```python
from SAM import SAM
import matplotlib.pyplot as plt
import utilities as ut
import numpy as np

%matplotlib inline
```

Next, we initialize SAM and load our data and annotations file. Cells are all annotated by the time since their induction with LPS (75 minutes, 150 minutes, or 300 minutes).

```python
sam=SAM()
sam.load_data('final_datasets/GSE94383/GSE94383_d2.csv')
sam.load_obs_annotations('final_datasets/GSE94383/GSE94383_ann2.csv')
```

Now, preprocess the data and convert the annotations vector to strings (so that the time points show up as discrete labels in the scatter plot colorbars).

Here, we are using the `include_cells` argument of `preprocess_data()` to tell SAM to keep cells from time points that are NOT 0 minutes. In this dataset, Cells sampled at 0 minutes have either not been induced with LPS or only just been induced.

```python
cells_to_include = sam.adata.obs_names[sam.adata.obs['Time point']!=0]
sam.preprocess_data(include_cells = cells_to_include)
sam.adata.obs['Time point'] = sam.adata.obs['Time point'].astype('<U3') #convert annotations from integers to strings
```

Finally, we run SAM with default parameters.

```python
sam.run()
```

Visualizing the UMAP projection, we find two clusters with the time points mixed into each other. 

```python
sam.scatter(c='Time point')
```

If we look at the expression pattern of Top2a, a canonical cell cycle marker, we find that the analysis is currently dominated by cell cycle effects.

```python
sam.show_gene_expression('Top2a',cmap='Spectral_r');
```

We use the `find_corr_genes` function in the `utilities.py` module to rank all genes based on their spatial correlation with Top2a. We take the top 200 of these genes as representatives of this cell cycle effect.

```python
cc_genes = ut.find_corr_genes(sam,'Top2a')[:200]
```

Here are the expression patterns of the top 5 genes. Indeed, they all have the same expression pattern as Top2a.

```python
fig, axs = plt.subplots(1,5)
fig.set_size_inches((24,4))
for i in range(len(axs)):
    sam.show_gene_expression(cc_genes[i],cmap='Spectral_r',axes = axs[i]);
```

Which principal components are dominated by cell cycle effects? If we can figure this out, we can subtract out these components from our dataset and rerun the SAM analysis. We can use `calculate_regression_PCs` to calculate the average loadings of the representative cell cycle genes on each principal component. Below, we calculate these scores for the top 100 PCs.

```python
scores = sam.calculate_regression_PCs(genes = cc_genes, npcs = 100)
plt.figure(); plt.plot(scores);
```

Visually, we can see that principal components 0, 1, and 8 have the highest scores. We can now use `regress_genes` to subtract out these principal components from the data. Note that `regress_genes` will only work after running `calculate_regression_PCs`.

```python
sam.regress_genes([0,1,8])
```

`sam.adata.X`, the matrix used for dimensionality reduction, now contains the dataset in which cell cycle effects have been removed. `sam.adata.layers['X_disp']`, the matrix used for calculating the spatial dispersions, is left unchanged (dispersion is ill-defined for negative values). It's no problem to use the original data for calculating dispersions, since the cell cycle gene expression patterns will not be reflected in the manifold after dimensionality reduction anyway, and so they will all receive low weights anyway.

We can now run SAM again with default parameters.

```python
sam.run()
```

Visualizing the resulting manifold, we can see two clusters with the time points properly ordered in each from 75 to 150 to 300 minutes.

```python
sam.scatter(c='Time point')
```

If we look at the cell cycle genes from before, we can see their expression patterns seem to be randomly distributed along the manifold, as we intended.

```python
fig, axs = plt.subplots(1,5)
fig.set_size_inches((24,4))
for i in range(len(axs)):
    sam.show_gene_expression(cc_genes[i],cmap='Spectral_r',axes = axs[i]);
```

As discussed in the paper, the two clusters are now separated by differences in expression patterns of genes that are central to various immunological pathways (like Tnf or Cxcl10).

```python
fig, axs = plt.subplots(1,2)
fig.set_size_inches((10,4))
sam.show_gene_expression('Tnf', cmap = 'Spectral_r', axes=axs[0]);
sam.show_gene_expression('Cxcl10', cmap = 'Spectral_r', axes=axs[1]);
```
